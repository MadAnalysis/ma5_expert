# $\LaTeX$ output example

```python
import ma5_expert as ma5
```

```python
sample = ma5.cutflow.Collection(
    "mass1000005_300.0_mass1000022_60.0_mass1000023_250.0_xs_5.689/Output/SAF/defaultset/atlas_susy_2018_31/Cutflows",
    xsection=5.689, 
    lumi=139.
)
```

```python
SRA_presel = [319.7,230.5,192.3,87.9,45.1,20.9,19.3,18.2,17.6,15.0,13.7]

ATLAS = ma5.cutflow.Collection() 

ATLAS.addSignalRegion('SRA',   sample.SRA.CutNames,   SRA_presel+[13.7])
ATLAS.addSignalRegion('SRA_L', sample.SRA_L.CutNames, SRA_presel+[0.4])
ATLAS.addSignalRegion('SRA_M', sample.SRA_M.CutNames, SRA_presel+[6.4])
ATLAS.addSignalRegion('SRA_H', sample.SRA_H.CutNames, SRA_presel+[7.0])
```

`write_comparison_table` simply writes a comparisson table between multiple samples designed for [PAD](http://madanalysis.irmp.ucl.ac.be/wiki/PublicAnalysisDatabase) validation sequence

```python
ma5.cutflow.CutFlowTable(ATLAS,sample, 
                         notes=r' Some notes here for the caption ',
                         sample_names=['ATLAS', '{\sc Ma5 - SFS}']).write_comparison_table()
```

`write_signal_comparison_table` writes a signal vs bacground comparisson table

```python
ma5.cutflow.CutFlowTable(ATLAS,ma5, 
                         notes=r' Some notes here for the caption ',
                         sample_names=['ATLAS', '{\sc Ma5 - SFS}']).write_signal_comparison_table()
```
